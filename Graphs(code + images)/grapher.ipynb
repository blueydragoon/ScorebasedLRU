```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
data1 = pd.read_csv('RECE1.csv')
data2 = pd.read_csv('GEO1.csv')
data3 = pd.read_csv('LRU1.csv')
data4 = pd.read_csv('LRFU1.csv')
data5 = pd.read_csv('SW1.csv')

# Plot
plt.figure(figsize=(10,10))
plt.plot(data1['size'], data1[' hitRatio'],label='RECE1')
plt.plot(data2['size'], data2[' hitRatio'],label='GEO1')
plt.plot(data3['size'], data3[' hitRatio'],label='LRU1')
plt.plot(data4['size'], data4[' hitRatio'],label='LRFU1')
plt.plot(data5['size'], data5[' hitRatio'],label='SW1')
#plt.xticks(data['size'], data[' hits'])
plt.legend()
plt.xlabel('size')
plt.ylabel('hitRatio')
#plt.show()
plt.savefig('COM1.png')
```
